### Задача 1
Ймовірність того, що при перевірці деталь виявиться стандартною, дорівнює 0,6. Перевірці підлягають три деталі. Побудувати закон розподілу двох дискретних випадкових величин X - появи числа бракованих деталей і Y - появи числа стандартних деталей. Обчислити коефіцієнт кореляції $r_xy$.

### Рішення.

Побудуємо таблиці розподілу випадкових величин X,Y:

Маємо 4 можливі випадки виникнення подій X та Y. Кількість бракованих делатей може бути 0, 1, 2 або 3 це також вірно для стандартних деталей.

За умовами задачі:
$P(X_i)=0.4 \,\,\,\,\, P(X_i^-)=0.6 \,\,\,\,\,  P(Y_i)=0.6 \,\,\,\,\, P(Y_i^-)=0.4$

Для випадку коли всі деталі браковані:

$P(X=0)=P(X_1^-*X_2^-*X_3^-)=P(X_1^-)*P(X_2^-)*P(X_3^-)=0.6*0.6*0.6=0.216$ 

Для випадку коли одна деталь бракована інші стандартні:

$P(X=1)=P(X_1)*P(X_2^-)*P(X_3^-)+P(X_1^-)*P(X_2)*P(X_3^-)+P(X_1^-)*P(X_2^-)*P(X_3)=0.4*0.6*0.6+0.6*0.4*0.6+0.6*0.4*0.6=0.432$

Для випадку коли дві деталі браковані одна стандартна:

$P(X=2)=P(X_1)*P(X_2)*P(X_3^-)+P(X_1^-)*P(X_2)*P(X_3)+P(X_1)*P(X_2^-)*P(X_3)=0.4*0.4*0.6+0.6*0.4*0.4+0.4*0.4*0.6=0.288$

Для випадку коли три деталі браковані:

$P(X=3)=P(X_1)*P(X_2)*P(X_3)=0.4*0.4*0.4=0.064$

Отримали ряд розподілення для X: 

|$X_i$|0|1|2|3|
|-|-|-|-|-|
|$P(X_i)$|0.216|0.432|0.288|0.064|

Аналогічно для Y:

$P(Y=0)=P(Y_1^-*Y_2^-*Y_3^-)=P(Y_1^-)*P(Y_2^-)*P(Y_3^-)=0.4*0.4*0.4=0.064$ 

Для випадку коли одна деталь стандартна інші браковані:

$P(Y=1)=P(Y_1)*P(Y_2^-)*P(Y_3^-)+P(Y_1^-)*P(Y_2)*P(Y_3^-)+P(Y_1^-)*P(Y_2^-)*P(Y_3)=0.6*0.4*0.4+0.4*0.6*0.4+0.4*0.4*0.6=0.288$

Для випадку коли дві деталі стандартні одна бракована:

$P(Y=2)=P(Y_1)*P(Y_2)*P(Y_3^-)+P(Y_1^-)*P(Y_2)*P(Y_3)+P(Y_1)*P(Y_2^-)*P(Y_3)=0.6*0.6*0.4+0.4*0.6*0.6+0.6*0.4*0.6=0.432$

Для випадку коли три деталі стандартні:

$P(Y=3)=P(Y_1)*P(Y_2)*P(Y_3)=0.6*0.6*0.6=0.216$

Отримали ряд розподілення для Y: 

|$Y_i$|0|1|2|3|
|-|-|-|-|-|
|$P(Y_i)$|0.064|0.288|0.432|0.216|


Вирахуємо математичні очіукування $M(X),M(Y),M(X,Y)$:

$$M(X)= \sum_{i=0}^{3}X_i P_i =0*0.216+1*0.432+2*0.288+3*0.064=1.2$$
$$M(Y)= \sum_{i=0}^{3}Y_i P_i =0*0.064+1*0.288+2*0.432+3*0.216=1.8$$
$$M(X,Y)=\sum_{i=0}^{3} \sum_{j=0}^{3} Y_iX_jP_{ij}$$

Розрахуємо $P_{ij}$ за допомогою коду Python


```python
x=[0,1,2,3]
y=[0,1,2,3]
px=[0.216,0.432,0.288,0.064]
py=[0.064,0.288,0.432,0.216]
pxy=[]
st=[]
for j in range(4):
    st=[]
    for i in range(4):
        st.append(px[i]*py[j])
    pxy.append(st)
    
for j in range(4):
    for i in range(3):
        print('P(',j,',',i,')=','{0:10.6f}  '.format(pxy[j][i]),end="") 
    print('P(',j,',',3,')=','  {0:10.6f}  '.format(pxy[j][3]))
    
```

Тепер проведемо остаточні обрахунки за формулами.

$$D(X)=\sum_{i=1}^{3}X_i^2 P_i - M(X)^2 \,\,\,\,\,\,\,D(Y)=\sum_{i=1}^{3}Y_i^2 P_i - M(Y)^2 $$

$$cov(X,Y)=M(X,Y)-M(X)M(Y) \,\,\,\,\,\,\,\,\,\,\,\,\, r_{xy}=\frac{cov(X,Y)}{\sqrt{D(X)D(Y)}}  $$

```python
mxy=0  # Математичні очікування
mx=0
my=0
for i in range(4):
    for j in range(4):
        mxy+=y[i]*x[j]*pxy[i][j]

for i in range(4):
    mx+=x[i]*px[i]
    my+=y[i]*py[i]
print('Математичне очікування')
print('Mx={0:8.5f}'.format(mx),'My={0:8.5f}'.format(my),'Mxy={0:8.5f}'.format(mxy))

dx=0 #Дисперсії
dy=0
for i in range(4):
    dx+=x[i]*x[i]*px[i]
    dy+=y[i]*y[i]*py[i]
dx=dx-mx*mx
dy=dy-my*my
print('Дисперсія')
print('DX={0:8.5f}'.format(dx),"   DY={0:8.5f}".format(dy))
cov=mxy-mx*my
print("Коваріація  cov(xy)={0:8.5f}".format(cov))
rxy=cov/(dx*dy)**0.5
print("Кореляція  Rxy={0:8.5f}".format(rxy))
```
