###  Задача 3
Два верстати-автомати виготовляють однакові вироби, які попадають на загальний конвеєр. Продуктивність першого автомата у два рази більше ніж продуктивність другого. Перший верстат-автомат виготовляє 60% виробів вищого гатунку, а другий 84%. А) Знайти ймовірність того, що навмання взятий  з конвеєра вибір вищого гатунку. В) Знайти ймовірність того, що він виготовлений на першому верстаті.

#### A)

### Розв'язання
Введемо групу гіпотез.
Нехай $ Н_1 $ це вироби які воготовляються на першому верстаті, тоді $ Н_2 $ - вироби виготовлені на другому верстаті.

Ймовірність гіпотез знайдемо як відношення кількості деталей вироблених кожним станком до загальної кількості виробів.

$P(H_1)= \frac {2}{3} \,\,\,\,\,$      відповідно -  $\,\,\,\,\,     P(H_2)= \frac {1}{3} $

Введемо поняття ймовірності події $A$. Тоді за умов задачі апріорні ймовірності отримати вироб вищого гатунку дорівнюють: 
$$ P(A|H_1)=0.6 \,\,\,\,\,\,\,  P(A|H_2)=0.84$$

Ймовірність події $А$ знайдемо по формулі повних ймовірностей 
$$ P(A)=\displaystyle\sum_{i=1}^{2}P(A|H_i) P(H_i)= 0.6 \frac {2}{3} +0.84 \frac {1}{3}=0.68$$ 

### Відповідь
Ймовірність того, що навмання взятий вироб буде вищого гатунку дорівнює **0.68**
***

####  B)

Розглянемо ймовірність того, що вироб вищого гатунку належить першому верстату.

Для цього застосуємо формулу Байєса:

$$P(H_1|A)=\frac{P(H_1)*P(A|H_1)}{P(H_1)*P(A|H_1)+P(H_2)*P(A|H_2)}=\frac{\frac{2}{3}*0.6}{0.68} \approx0.5882$$


### Відповідь
Ймовірність того, що навмання взятий вироб буде вищого гатунку та вироблений на першому верстаті дорівнює **$\approx 0.5882$**
