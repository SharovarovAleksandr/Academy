### Задача 2
Для сигналізації про аварію встановлені два незалежно працюючих сигналізатори. Ймовірність того що під час аварії спрацює перший сигналізатор дорівнює 0.95, ця ймовірність для другого сигналізатора дорівнює 0.9. Знайти ймовірність того що при аварії спрацює:

а) хоча б один сигналізатор;

в) тільки один сигналізатор.

### Розв'язання
За умовами задачі маємо схему класичного паралельного з'єдання елементів схеми.

<img src="img1.jpg">

Загальна ймовірність роботи схеми становить суму ймовірностей роботи кожного елементу. $ A=A_1+A_2$

Розглянемо варіант а) - робота хоча б одного сигналізатора. Це твердження протилежно тому що не спрацівав жоден сигналізатор.
$ P=1-P(A^-)$
Ймовірніть того, що не спрацює сигналізатор $А_1$ дорівнює $P(A_1^-)=1-P(A_1)=1-0.95=0.05$

Аналогічно для $A_2: \,\,\,\,\,P(A_2^-)=1-P(A_2)=1-0.9=0.1$   

Ймовірність того, що не спрацюють одночасно два сигналізатора $P(A^-)=P(A_1^-)P(A_2^-)=0.05*0.1=0.005$

Ймовірність того що спрацює хоча б один $P=1-P(A^-)=1-0.005=0.995$
*** 

Розглянемо варіант в) - спрацює тільки один сигналізатор.

Маємо дві ймовірності - 1) спрацює перший $P(A_1)=0.95$ та не спрацює другий $P(A_2^-)=0.1$ 

та навпаки - 2)  спрацює другий $P(A_2)=0.9$ та не спрацює перший $P(A_1^-)=0.05$

Загальна ймовірність того що спрацює тільки один дорівнює - $P(A)=P(A_1)P(A_2^-)+P(A_2)P(A_1^-)=0.95*0.1+0.9*0.05=0.14$

### Відповідь
а) Ймовірність спраціювання хоча б одного становить __0,995__

в) Ймовірність що спрацює тільки один - __0,14__
