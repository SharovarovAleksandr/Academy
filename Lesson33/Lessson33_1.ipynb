```python
import matplotlib as ml
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
import seaborn as sb
```

### Завдання 1
Загрузка DataFrame з файлу csv.

```python
df=pd.read_csv('cars1.csv')
df
```

### Завдання 2
Візуалізація залежностей. Fuel-Cost, Fuel-Speed, Speed-Load_capacity.

```python
df.sort_values(by=['Fuel'], inplace=True) #Для наглідності результатів проведемо попереднє сортування за параметром Fuel
fig=plt.figure(figsize=(15,5), dpi=80)
ax1=fig.add_subplot(1,3,1)
ax1.set_title('Fuel-Cost relation')
ax1.set_xlabel('Fuel')
ax1.set_ylabel('Cost')
ax1.plot(df['Fuel'],df['Cost'],color='Green')

ax2=fig.add_subplot(1,3,2)
ax2.set_title('Fuel-Speed relation')
ax2.set_xlabel('Fuel')
ax2.set_ylabel('Speed')
ax2.scatter(df['Fuel'],df['Speed'],color='Red')

ax3=fig.add_subplot(1,3,3)
ax3.set_title('Speed-Load_capacity relation')
ax3.set_xlabel('Speed')
ax3.set_ylabel('Load_capacity')
ax3.scatter(df['Speed'],df['Load_capacity'],color='Blue')
```

### Завдання 3
Побудова гістограми для розділення автомобілів на 5 груп за швидкістю.

```python
fig=plt.figure(figsize=(9,5), dpi=80)
ax=fig.add_subplot()
ax.set_title('Histogram 5 groups speed')
ax.set_xlabel('Groups')
ax.set_ylabel('Speed')
ax.hist(df['Speed'],5,width=25)
```

### Завдання 4
Коробковий графік для відображення швидкостей всіх марок.

```python
fig=plt.figure(figsize=(8,6), dpi=80)
ax=fig.add_subplot()
ax.set_title('Box plot speed cars')
ax.set_xlabel('Cars')
ax.set_ylabel('Speed')
ax.boxplot(df['Speed'])
```

##### Коробковий графік відображення скоростей авто за іх типами (Seaborn)

```python
sb.boxplot(x="Type", y="Speed", data=df)
```
