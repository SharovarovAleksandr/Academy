### Задача 2
Є можливість вибрати спосіб виробництва і реалізації двох наборів товарів широкого вжитку. За даними відділу маркетингу, яким були проведені дослідження ринку, можливий прибуток від виробництва і реалізації Х і Y наведено в таблицях (Х, Y – прибуток у грошових одиницях):

|X|1000|1500|2000|  
|--|----|----|----|         
|P|0.5|0.3|0.2| 

відповідно

|Y|1000|1500|1750|
|--|----|----|----|
|P|0.4|0.4|0.2|

Потрібно оцінити ступінь ризику і запропонувати рішення щодо випуску і реалізації одного із наборів товарів.

### Розв'язання
Знаходимо математичне очікування для кожного варіанту $M(X)$ та $M(Y)$

$$M(X)=\sum_{i=1}^{3}X_i P_i= 1000*0.5+1500*0.3+2000*0.2=1350$$ $$ M(Y)=\sum_{i=1}^{3}Y_i P_i=1000*0.4+1500*0.4+1750*0.2=1350$$

Вирахуємо дисперсію кожного варіанту:

$$D(X)=\sum_{i=1}^{3}X_i^2 P_i - (\sum_{i=1}^{3}X_i P_i)^2=1000^2*0.5+1500^2*0.3+2000^2*0.2-1350^2=152500$$
$$D(Y)=\sum_{i=1}^{3}Y_i^2 P_i - (\sum_{i=1}^{3}Y_i P_i)^2=1000^2*0.4+1500^2*0.4+1750^2*0.2-1350^2=90000$$

Вирахуємо середньоквадратичне відхилення для X та Y:
$$ \sigma(X)=\sqrt{D(X)} =\sqrt{152500}\approx{390.5125} $$
$$ \sigma(Y)=\sqrt{D(Y)} =\sqrt{90000}=300 $$
$$\sigma(X)>\sigma(Y)$$

### Відповідь
Ризик пов'язаний із випуском продукції $Х$ вищий ніж $Y$
