### Задача 1
Два економісти заповнюють документи з власними прогнозами щодо курсу певних коштовних паперів, після чого їх складають у спільну папку. Ймовірність відхилення прогнозованого значення курсу від фактичного понад допустиму похибку для першого економіста дорівнює 0,1; для другого – 0,2. Перший економіст заповнив 40 документів, другий – 60. Під час перевірки навмання взятий із папки документ виявився з суттєвою похибкою. Визначити ймовірність того, що його склав перший економіст.

### Розв'язання

Введемо пойняття події $E_1$ - документ заповнив перший економіст та $Е_2$ - документ заповнив другий економіст.

Тоді: $$P(E_1)=\frac{40}{100}=0.4  \,\,\,\,\, відповідно, \,\,\,\,\,P(E_2)=\frac{60}{100}=0.6$$

За умовами задачі   $P(A|E_1)=0.1 \,\, та \,\,P(A|E_2)=0.2$

Застосуємо формулу повної ймовірності - 

$$P(A)=P(E_1)*P(A|E_1)+P(E_2)*P(A|E_2)=0.4*0.1+0.6*0.2=0.16$$

Відповідно $P(E_1|A)$ за формулою Бейтса:

$$P(E_1|A)=\frac{P(E_1)*P(A|E_1)}{P(A)}=\frac{0.4*0.1}{0.16}=0.25$$


### Відповідь
Ймовірність того, що документ склав перший економіст дорівнює __0,25__
