### Завдання
#### Визначьте точки відповідності на зображеннях chess_1 та chess_2. Проведіть гомографічне перетворення для приведення зображення chess_2 до вигляду базового зображення chess_1.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv2
from google.colab.patches import cv2_imshow
```

```python
img1= cv2.imread("/content/drive/MyDrive/Colab Notebooks/chess_1.png")
cv2_imshow(img1)
```

```python
img2= cv2.imread("/content/drive/MyDrive/Colab Notebooks/chess_2.png")
cv2_imshow(img2)
```

```python
pts_src = np.array([ [192, 51], [463, 37], [198, 246],[460, 262] ])

pts_dst = np.array([[189, 317],[164, 34],[408, 365],[530, 102]])
```

### Перетворення chess_1 до chess_2

```python
h, status = cv2.findHomography(pts_src, pts_dst)
h
```

```python
im_out = cv2.warpPerspective(img1, h, (img1.shape[1], img1.shape[0]))
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```

### Перетворення chess_2 до chess_1

```python
h, status = cv2.findHomography(pts_dst, pts_src)
im_out = cv2.warpPerspective(img2, h, (img2.shape[1], img2.shape[0]))
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```
