```python
import pandas as pd
import openpyxl

# Створення DataFrame за допомогою завантаження файлу exel
file="students.xlsx"
ex=pd.read_excel(file,sheet_name='groupe_1')
ex
```

```python
# Додавання колонки до DataFrame за допомогою Series
allowance=pd.Series([True,False,True,True,False,True,True,False,True,False,True,True,False,True,True])
ex['allowance']=allowance
ex
```

```python
# Аналіз колонки середній бал студентів
ex.agg({'average_grade' : ['min', 'mean','median','max']})
```

```python
# Спосіб 1
# Підрахунок кількості студентів що здали екзамен з першого разу 
ex_filter=ex.query('number_attempts==1')
ex_filter['number_attempts'].count()
```

```python
# Спосіб 2
# Підрахунок кількості студентів що здали екзамен з першого разу 
len(ex[ex['number_attempts']==1])
```

```python
# Створення Series з DataFrame
sr=pd.Series(dtype='int64')
sr=ex['month_grade']
sr
```

```python
#  Запис Series у файл csv
sr.to_csv('series.csv')
```
