Демонстрація роботи із магічними командами __%%time__ та __%%timein__. В данному випадку команда  __%%time__ буде вираховувати загальний час виконання завдання, тоді як, команда  __%%timein__ буде виконувати це завдання 10 разів та вираховувати середні показники.

```python
%%time
%%timeit
import random

def rand_int(n,start,end):
    l=[]
    for i in range(n):
        l.append(random.randint(start,end))
    return l

def rand_float(n,start,end):
    l=[]
    for i in range(n):
        l.append(random.uniform(0.1,100))
    return l
s=0
int_list=rand_int(50000,0,1000)
for i in int_list:
    s+=i
print(f'Кількість випадково згенерованих цілих чисел {len(int_list)} їх сума дорівнює {s}')
float_list=rand_float(50000,0,1000)
for i in float_list:
    s+=i   
print(f'Кількість випадково згенерованих чисел {len(float_list)} їх сума дорівнює {s}')
```
