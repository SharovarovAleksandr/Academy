Побудуємо функцію F*(x)
   
$$ F^*(x) = W(X<x)=\frac{n_x}{n}= \begin{cases}
  0,  & x\leq  6 \\
  0.025  & 6 <x\leq  7 \\
  0.1 & 7<x \leq  8  \\
  0.25 & 8 <x\leq 9  \\
  0.45 & 9 <x\leq 10  \\
  0.6 & 10 <x\leq  11  \\
  0.75 & 11 <x\leq  12  \\
  0.875 & 12<x \leq 13  \\
  0.95 & 13 <x\leq 14  \\
  1 & x>14   \\
  \end{cases}
$$
   
   Модою даної виборки є число $M_0=9$ , тому що воно зустрічається 8 разів, а медіаною $M_e=10$
