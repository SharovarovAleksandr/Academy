### Задача 2
Протягом 30 днів вересня було 12 дощових, 8 - вітряних, 4- холодних, 5 - дощових та вітрених, 3- дощових та холодних, 2- вітряних та холодних, а один день був вітряним, дощовим та холодним одночасно. Скількі днів у вересні була гарна погода?

### Намалюємо відповідну діаграму із використанням бібліотеки matplotlib_venn


```python
from matplotlib_venn import venn3
from matplotlib import pyplot as plt

venn3(subsets=(12, 8, 5, 4, 3, 2, 1), 
      set_labels=('Rain', 'Wind','Cool'), 
      set_colors=("orange", "blue", "red"), alpha=0.7)
plt.show()
```

### Розв'язання
Нехай |R| - потужність множини кількості днів із дощем. Відповідно: |W| - вітряних, |C| - холодних.

За умовами задачі: |R| = 12 |W|=8 |C|=4 |R ∩ W|=5 |W ∩ C|=2 |R ∩ C|=3 |R U W U C|=1

Скористаємося формулою множин та з'ясуємо скільки днів у місяці були непогодними :

|R U W U C|=|R| + |W| + |C| - |R ∩ W| - |W ∩ C| - |R ∩ C| + |R ∩ W ∩ C|

|R U W U C|=12+8+4-5-2-3+1=15

Відповідно, кількість днів із гарною погодою дорівнює кількості днів місяця мінус непогодні дні.
### 30-15=15

### Відповідь 15 днів.
