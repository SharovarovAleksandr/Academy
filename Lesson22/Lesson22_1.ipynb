### Задача 1
За підсумками іспитів із 37 студентів відмінну оцінку з математики мали 15 студентів, із фізики - 16, із хімії - 19, з математики та фізики - 7, з математики та хімії - 9, з фізиик та хімії - 6, з у сіх трьох дисциплін - 4. Скільки студентів отримали хоча б по одній відмінній оцінці?

### Намалюємо відповідну діаграму із використанням бібліотеки matplotlib_venn

```python
from matplotlib_venn import venn3
from matplotlib import pyplot as plt

venn3(subsets=(15, 16, 7, 19, 9, 6, 4), 
      set_labels=('Math', 'Phisic','Chemistry'), 
      set_colors=("orange", "blue", "red"), alpha=0.7)
plt.show()
```

### Розв'язання
Нехай |M| - потужність множини студенів, які отримали відмінну оцінку з математики. 
Відповідно: |P| - з фізики, |C| - з хімії.

За умовами задачі: |M| = 15  |P|=16   |C|=19   |M ∩ P|=7   |M ∩ C|=9   |P ∩ C|=6   |M U P U C|=4

Скористаємося формулою множин:

|M U P U C|=|M| + |P| + |C| - |M ∩ P| - |M ∩ C| - |P ∩ C| + |M ∩ P ∩ C|

|M U P U C|=15+16+19-7-9-6+4=32

### Відповідь - 32 студента отримали хоча б по одній відмінній оцінці.
