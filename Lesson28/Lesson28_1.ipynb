### Задача 1
Із генеральної сукупності, розподіленої за нормальним законом, отримано вибірку об’ємом n=20

|$x_i$|1|2|4|6|7|
|--|-|-|-|-|-|
|$n_i$|2|3|10|3|2|

Знайти з надійністю $\gamma=0,95$ інтервальну оцінку для середнього квадратичного відхилення $\sigma$ генеральної сукупності


### Рішення

Довірчі інтервали для оцінки середнього квадратичного відхилення $\sigma$ нормально розподіленої ознаки x з надійністю $\gamma$ за виправленим вибірковим середнім квадратичним відхиленням $s$ знаходяться за формулами:
 

$$ s(1-q)< \sigma< s(1+q),  \,\,\,\, для \,\,q<1;$$


$$ 0< \sigma <s(1+q),\,\,\,\, для \,\,q>1 $$

Обчислемо вибіркове середне:

$$ \bar{x_b}=\frac{1}{n}\sum_{i=1}^{5}x_in_i = \frac{(1*2+2*3+4*10+6*3+7*2)}{20}=4   $$

Знайдемо виправлену дисперсію:

$$ s^2=\frac{1}{n-1}\sum_{i=1}^{5}(x_i-\bar{x_b})^2n_i= \frac{(1-4)^2*2+(2-4)^2*3+(4-4)^2*10+(6-4)^2*3+(7-4)^2*2}{19}\approx3.1579$$

Виправлене середньоквадратичне відхилення $s=\sqrt{s^2}=\sqrt{3.1579}\approx 1.777$

Із таблиці знаходемо $q=q(\gamma,n)=q(0.95,20)=0.37$  виходячи з того що, $q<1$ використовуємо формулу $ s(1-q)< \sigma< s(1+q)$

$$1.777(1-0.37)<\sigma<1.777(1+0.37)$$
$$1.11951<\sigma<2.43449$$


### Відповідь
Довірчий інтервал становить (1.11951;2.43449)
