1. Використовуючи бібліотеку gzip, напишіть код для стиснення текстового файлу.

```python
import lorem
import gzip
import os
import sys
```

```python
# Генеруємо текст
text= lorem.get_word(count=5000)
text
```

```python
# Стиснення файлу
with open('file1.txt', 'wb') as f:
   f.write(text.encode('utf-8'))
size = os.path.getsize("file1.txt")
f.close
print("File size = ",size)
with open('file1.txt', 'rb') as f:
    with gzip.open('file1.txt.gz', 'wb') as f_out:
        f_out.write(f.read())
size1 = os.path.getsize("file1.txt.gz")
print('Стиснення файлу завершено.')
print("File gz size = ",size)
print("Коефіцієнт стиснення = ", size/size1)
```

```python
# Стиснення тексту
size = sys.getsizeof(text)
print("Розмір тексту = ", size, "байт")
with gzip.open('file.gz', 'wb') as f:
    f.write(text.encode('utf-8'))
print('Стиснення файлу завершено.')
size1 = os.path.getsize("file.gz")
print(f'Розмір стиснутого файлу : {size1} байт')
print("Коефіцієнт стиснення = ", size/size1)
```
