### Задача

Протягом 50 років вимірювався рівень води навесні під час повені відносно умовного нуля. Результати вимірювання наведено у вигляді інтервального статистичного розподілу:

|$x_i$|0-24|24-48|48-72|72-96|96-120|120-144|144-168|168-192|192-216|
|---|----|-----|-----|-----|------|-------|-------|-------|-------|
|$n_i$|1|2|4|6|12|16|6|2|1|

За даними статистичними розподілами вибірки висунути $H_0$ про закон розподілу ознаки генеральної сукупності і при рівні значущості $\alpha=0.01$ перевірити її привильність.

### Рішення

Обчислимо теоретичні частоти, а для цього знайдемо значення $\bar{x_B}$ та $\sigma_B$ побудувавши дискретний розподіл за заданим інтервальним, а саме:

|$x_i$|12|36|60|84|108|132|156|180|204|
|-----|--|--|--|--|---|---|---|---|---|
|$n_i$|1|2|4|6|12|16|6|2|1|

Розмір вибірки дорівнює $$n=\sum_{i=1}^{9}n_i=1+2+4+6+12+16+6+2+1=50$$


$$ \bar{x_b}=\frac{1}{n}\sum_{i=1}^{9}x_in_i =\frac{12*1+36*2+60*4+84*6+108*12+132*16+156*6+180*2+204*1}{50}=\frac{5736}{50}=114.72$$

$$D_B=\frac{ \sum x_i^2n_i }{n}-( \bar{x_b})^2=\frac{12^2*1+36^2*2+60^2*4+84^2*6+108^2*12+132^2*16+156^2*6+180^2*2+204^2*1}{50}-(114.72)^2=\frac{730656}{50}-13160.6784=1452.4416$$

$\sigma_B=\sqrt{D_B}=\sqrt{1452.4416}\approx38.1109 \,\,\,\,$ Зробимо обчислення теоретичних частот у наступній таблиці:


$Z_i=\frac{x_i-\bar{x_b}}{\sigma_B};\,\,\,\,\, Z_{i+1}=\frac{x_{i+1}-\bar{x_b}}{\sigma_B};\,\,\,\,\,  n_i^`=n(\Phi(Z_{i+1})-\Phi(Z_i))$
